```python
icd_df = pd.read_csv('DIAGNOSES_ICD.csv')
```

```python
icd_df.describe()
```

```python
print(icd_df.isnull().sum().sum())
icd_df.dropna(how='any', inplace=True)
print(icd_df.isnull().sum().sum())
```

```python
def convert_icd9(icd9_object):
  icd9_str = str(icd9_object)
  if icd9_str[0] == 'E':
    converted = icd9_str[0:4]
  else:
    converted = icd9_str[0:3]
  return converted


new_df1 = icd_df[['ICD9_CODE']]
new_df1['ICD9_CODE'] = new_df1['ICD9_CODE'].apply(convert_icd9)
```

```python
len(pd.unique(new_df1['ICD9_CODE']))
```

```python
new_df = new_df1.groupby(['ICD9_CODE'])['ICD9_CODE'].count()
new_df
```

```python
new_df.plot()
```

```python
new_df1['ICD9_CODE'].value_counts()
```

```python
new_df1['freq'] = new_df1.groupby(by='ICD9_CODE')['ICD9_CODE'].transform('count')
```
